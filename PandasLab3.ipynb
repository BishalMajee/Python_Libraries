```python
import pandas as pd
df=pd.read_csv(r"C:\Users\Bishal Majee\Desktop\panda\sample data.txt")
df
```

```python
missing_values = df.isna().sum()
missing_values
```

```python
df=df.fillna('')
df
```

```python
df['PHONE_NUMBER'] = df['PHONE_NUMBER'].str.replace('-', '', regex=False)
df['PHONE_NUMBER']
```

```python
df=df.drop_duplicates()
df
```

```python
df['HIRE_DATE'] = pd.to_datetime(df['HIRE_DATE'])
```

```python
df.info()
```

```python

df['SALARY'] = pd.to_numeric(df['SALARY'], errors='coerce')

# Handle missing values (e.g., replace NaN with 0 or remove rows with NaN)
df['SALARY'] = df['SALARY'].fillna(0).astype(int)

outliers = df[(df['SALARY'] < 30000) | (df['SALARY'] > 200000)]

print("DataFrame after cleaning and conversion:\n", df)
print("\nOutliers:\n", outliers)
df
```

```python
median_salary = df['SALARY'].median()
median_salary

```

```python
df['SALARY'] = df['SALARY'].apply(lambda x: median_salary if x < 30000 or x > 200000 else x)
print(df['SALARY'])
```

```python
df['FULL_NAME'] = df['FIRST_NAME'] + ' ' + df['LAST_NAME']
df
```

```python
filename='Sample data.csv'
df.to_csv(filename)
```
