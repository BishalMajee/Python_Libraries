### 1.Using SciPy Linear Algebra please solve the below problem.

 Input: 

7x + 2y = 8

4x + 5y = 10

```python
import numpy as np
from scipy.linalg import solve


A = np.array([[7, 2],
              [4, 5]])


b = np.array([8, 10])


solution = solve(A, b)
solution
```

### 2.Using SciPy Linear Algebra please solve the below problem.

 Input:

 x + 3y +10z = 10 

2x + 12y + 7z = 18 

5x + 8y + 8z = 30

```python
import numpy as np
from scipy import linalg

A = np.array([[1, 3, 10],
              [2, 12, 7],
              [5, 8, 8]])


b = np.array([10, 18, 30])


solution = linalg.solve(A, b)

print(f"x = {solution[0]}, y = {solution[1]}, z = {solution[2]}")
```

### Using a Sobel Filter, filter the image and then display it.

```python
import scipy.ndimage as nd
import matplotlib.pyplot as plt
import imageio
img_var= imageio.imread(r'C:\Users\Bishal Majee\Downloads\dog.jpg')

sobel_filter = nd.sobel(img_var)


plt.imshow(sobel_filter, cmap='gray')
plt.axis('off')
plt.title('Sobel Filter')
plt.show()
```
