```python
import pandas as pd
from scipy.constants import pi
df=pd.read_csv(r'https://raw.githubusercontent.com/AnudipAE/DANLC/master/radius_data.csv')
```

```python
df.head()
```

```python
df['Area']=df['Radius']**2*pi
```

```python
df.head()
```

```python
df.to_csv('new_kilometer addeed',index=False)
```

```python
df=pd.read_csv(r'https://raw.githubusercontent.com/AnudipAE/DANLC/master/city_temperatures.csv')
df
```

```python
from scipy.constants import convert_temperature
df['Temperature (°F)']=convert_temperature(df['Temperature (°C)'],'C','F')
```

```python
df
```

```python
df.to_csv('Fahrenheit addeed',index=False)
```

```python
import pandas as pd
df=pd.read_csv(r'https://raw.githubusercontent.com/AnudipAE/DANLC/master/people_heights.csv')
df.head()
```
